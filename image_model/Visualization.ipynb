```python
from keras.datasets import fashion_mnist, mnist
from sklearn.decomposition import PCA
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
(Xs_tr, Ys_tr), (Xs_te, Ys_te) = mnist.load_data()

Xs_tr = Xs_tr.reshape(Xs_tr.shape[0], Xs_tr.shape[1] * Xs_tr.shape[2])
Xs_tr = PCA(n_components=2).fit_transform(Xs_tr)

plt.figure(figsize=(20,20))
colors = ['b', 'g', 'r', 'c', 'm','y','plum','darkorange','slategray','salmon']
target_names = np.unique(Ys_tr)
lw = 2
for color, i, target_name in zip(colors, target_names, target_names):
 plt.scatter(Xs_tr[Ys_tr == i, 0], Xs_tr[Ys_tr == i, 1], color=color, alpha=.8, lw=lw,
 label=target_name, s=1)
plt.legend(loc='best', shadow=False, scatterpoints=1)
plt.title('PCA of Fashion MNIST dataset')
plt.show()
```
